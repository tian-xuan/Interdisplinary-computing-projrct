```python
def solution(x, outcome):
    a=np.matrix(x)
    inv=np.linalg.pinv(a)
    sol=np.matrix(outcome).transpose()
    weights=np.matmul(inv,sol)
    print(weights)
    predicted=np.matmul(a,weights).tolist()
    return predicted
predicted=solution(x,outcome)


```
